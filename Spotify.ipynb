```python
from datetime import date, datetime

spotify_premium_dates = []
spotify_premium_users = []

for line in spotify_premium.split("\n"):
    quarter, user = line.split("\t")
    q,year = quarter.split(" ")
    m = int(q.strip("Q"))*3-2
    d = date(day = 1, month=m, year=int(year))
    spotify_premium_dates.append(d)
    spotify_premium_users.append(float(user) * 1e6)
    
spotify_mau_dates = []
spotify_mau_users = []

spotify_dates_qarter = []
spotify_dates_year = []

for line in spotify_active_users.split("\n"):
    quarter, user = line.split("\t")
    q,year = quarter.split(" ")
    spotify_dates_qarter.append(int(q.strip("Q")))
    spotify_dates_year.append(int(year))
    m = int(q.strip("Q"))*3-2
    d = date(day = 1, month=m, year=int(year))
    spotify_mau_dates.append(d)
    spotify_mau_users.append(float(user) * 1e6)

spotify_rev_dates = []
spotify_rev_prem = []
spotify_rev_adv = []

for line in spotify_revenue_quarter.split("\n"):
    quarter, prem, adv = line.split("\t")
    q,year = quarter.split(" ")
    m = int(q.strip("Q"))*3-2
    d = date(day = 1, month=m, year=int(year))
    spotify_rev_dates.append(d)
    spotify_rev_prem.append(float(prem.replace(".","")) * 1e6)
    spotify_rev_adv.append(float(adv.replace(".","")) * 1e6)

adv_tot_dates = []
adv_tot_val = []
for line in adv_tot.split("\n"):
    d, adv = line.split("\t")
    d = datetime.strptime(d,"%d.%m.%Y").date()
    adv_tot_dates.append(d)
    adv_tot_val.append(float(adv.replace(".","")) * 1e6)


```

```python
# harmonise dates
import numpy as np
import pickle

spotify = {"date":[], "mau_users":[], "prem_users":[], "rev_prem": [], "rev_adv":[], "quarter": [], "year": [], "adv_tot":[]}

for d in spotify_mau_dates:
    if d in spotify_rev_dates:
        spotify["date"].append(np.datetime64(d))
        ind = spotify_rev_dates.index(d)
        spotify["rev_prem"].append(spotify_rev_prem[ind])
        spotify["rev_adv"].append(spotify_rev_adv[ind])
        ind = spotify_mau_dates.index(d)
        spotify["mau_users"].append(spotify_mau_users[ind])
        spotify["quarter"].append(spotify_dates_qarter[ind])
        spotify["year"].append(spotify_dates_year[ind])
        ind = spotify_premium_dates.index(d)
        spotify["prem_users"].append(spotify_premium_users[ind])
        ind = adv_tot_dates.index(d)
        spotify["adv_tot"].append(adv_tot_val[ind])
        

for k in spotify.keys():
    spotify[k] = np.asarray(spotify[k])
    
spotify["rev"] = spotify["rev_prem"] + spotify["rev_adv"]
spotify_norm = {}

for k in spotify.keys():
    if k == "date":
        spotify_norm[k] = (spotify[k]-spotify[k][0]).astype("float")
    else:
        spotify_norm[k] = spotify[k]/np.mean(spotify[k])

def gen_train_test():
    n = len(spotify["rev"])
    n_train_test = round(n*0.9)
    print(f"{n, n_train_test}")
    ind_all = np.arange(n)
    np.random.shuffle(ind_all)
    ind_train = ind_all[:n_train_test]
    ind_test = ind_all[n_train_test:]
    return [n, n_train_test, ind_train, ind_test]

    
def save_data(n, n_train_test, ind_train, ind_test):
    with open("spotify_data.pkl","wb") as f:
        pickle.dump([n, n_train_test, ind_train, ind_test],f)
        
def load_data():
    with open("spotify_data.pkl","rb") as f:
        return pickle.load(f)

try:
    [n, n_train_test, ind_train, ind_test] = load_data()
    print("loaded data")
except OSError as e:
    [n, n_train_test, ind_train, ind_test] = gen_data()
    save_data(n, n_train_test, ind_train, ind_test)
    print("generated new data, pleas regenerate the images")


print(ind_train, ind_test)
```
